```python
import numpy as np

# Define the function f(x1, x2)
def f(x1, x2):
    return np.array([np.sin(x1) * np.cos(x2), np.sin(x1) * np.sin(x2), np.cos(x1)])

# Define the function h(x1, x2)
def h(x1, x2):
    return x2 * np.sin(x1)

# Define the bounds for x1 and x2
bounds_x1 = (0, np.pi)
bounds_x2 = (0, 2 * np.pi)

# Number of divisions for Riemann integration
num_divisions_x1 = 1000
num_divisions_x2 = 1000

# Step size for each division
step_size_x1 = (bounds_x1[1] - bounds_x1[0]) / num_divisions_x1
step_size_x2 = (bounds_x2[1] - bounds_x2[0]) / num_divisions_x2

integral = 0
for i in range(num_divisions_x1):
    for j in range(num_divisions_x2):
        x1 = bounds_x1[0] + i * step_size_x1
        x2 = bounds_x2[0] + j * step_size_x2
        integral += h(x1, x2) * step_size_x1 * step_size_x2


print("Estimated Integral:", integral)
```
